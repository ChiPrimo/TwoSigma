```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 100)
```

```python
# Load news data file 1
news_train = pd.read_csv("news_train_df_1.csv") 
```

```python
news_train.head()
```

```python
# Convert all the 'time' to datetime format in news_train
news_train['time'] = pd.to_datetime(news_train['time']).apply(lambda x: x.date())
```

```python
# Generate a dataframe containing assetcodes and assetnames and subjects, audiences 
nlp_news = news_train[['time','assetCodes','assetName','headline','subjects','audiences' ]]
```

```python
nlp_news.head()
```

```python
# Test data in a small sample
#cutoff = pd.to_datetime("2007-01-01").date()
#df = nlp_news[~(NLP_news['time'] > cutoff)]
#df.head()
```

```python
# Group by assetName
#df_grouped = df.groupby('assetName').agg(''.join)
df_grouped = nlp_news.groupby('assetName').agg(''.join)
```

```python
len(df_grouped)
```

```python
df_grouped.head()
```

```python
# Get rid of the '' for assetCode, audiences
df_grouped['subjects'] = df_grouped['subjects'].apply(lambda x: x.replace("'",""))
df_grouped['audiences'] = df_grouped['audiences'].apply(lambda x: x.replace("'",""))
# Get rid of the {},'', and white space for subjects, audiences
df_grouped['subjects'] = df_grouped['subjects'].apply(lambda x: x.replace("{",""))
df_grouped['subjects'] = df_grouped['subjects'].apply(lambda x: x.replace("}",""))
df_grouped['subjects'] = df_grouped['subjects'].apply(lambda x: x.replace(" ",""))
df_grouped['audiences'] = df_grouped['audiences'].apply(lambda x: x.replace("{",""))
df_grouped['audiences'] = df_grouped['audiences'].apply(lambda x: x.replace("}",""))
df_grouped['audiences'] = df_grouped['audiences'].apply(lambda x: x.replace(" ",""))
```
